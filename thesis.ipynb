```python
from pybaseball import team_batting, team_pitching
import pandas as pd

years = range(2018, 2024)

all_data = []

for year in years:
    bat = team_batting(year)
    pit = team_pitching(year)

    bat["Season"] = year
    pit["Season"] = year

    # keep only needed columns (safe columns)
    bat = bat[["Team", "Season", "wOBA", "OPS", "wRC+"]]
    pit = pit[["Team", "Season", "FIP"]]

    df = pd.merge(bat, pit, on=["Team", "Season"])
    all_data.append(df)

fangraphs = pd.concat(all_data)

fangraphs.to_csv("fangraphs_features.csv", index=False)

print(fangraphs.head(180))
```

```python
from pybaseball import statcast
import pandas as pd
```

```python
months = [
    ("2023-03-01", "2023-03-31"),
    ("2023-04-01", "2023-04-30"),
    ("2023-05-01", "2023-05-31"),
    ("2023-06-01", "2023-06-30"),
    ("2023-07-01", "2023-07-31"),
    ("2023-08-01", "2023-08-31"),
    ("2023-09-01", "2023-09-30")
]

all_data = []

for start, end in months:
    print(f"Downloading {start} to {end}...")
    df = statcast(start_dt=start, end_dt=end)
    all_data.append(df)

statcast_2023 = pd.concat(all_data, ignore_index=True)

print(statcast_2023.shape)
```

```python
df = statcast_2023[[
    "game_date",
    "home_team",
    "away_team",
    "launch_speed",
    "launch_angle",
    "estimated_woba_using_speedangle"
]]
```

```python
df = df.dropna(subset=[
    "launch_speed",
    "launch_angle",
    "estimated_woba_using_speedangle"
])
```

```python
home_stats = df.groupby("home_team").agg({
    "launch_speed": "mean",
    "launch_angle": "mean",
    "estimated_woba_using_speedangle": "mean"
}).reset_index()

home_stats.columns = [
    "team",
    "avg_exit_velocity",
    "avg_launch_angle",
    "avg_xwOBA"
]
```

```python
away_stats = df.groupby("away_team").agg({
    "launch_speed": "mean",
    "launch_angle": "mean",
    "estimated_woba_using_speedangle": "mean"
}).reset_index()

away_stats.columns = [
    "team",
    "avg_exit_velocity",
    "avg_launch_angle",
    "avg_xwOBA"
]
```

```python
statcast_team = pd.concat([home_stats, away_stats])
statcast_team = statcast_team.groupby("team").mean().reset_index()

statcast_team["Season"] = 2023
```

```python
import pandas as pd

def process_retrosheet_file(file_path):
    # read raw file
    df_raw = pd.read_csv(file_path, header=None, encoding="latin1")

    # keep only the first 11 columns we need now
    basic_cols = [
        "date",                  # 0
        "doubleheader_code",     # 1
        "day_of_week",           # 2
        "visiting_team",         # 3
        "visiting_league",       # 4
        "visiting_game_num",     # 5
        "home_team",             # 6
        "home_league",           # 7
        "home_game_num",         # 8
        "visiting_score",        # 9
        "home_score"             # 10
    ]

    df = df_raw.iloc[:, :11].copy()
    df.columns = basic_cols

    # convert date
    df["date"] = pd.to_datetime(df["date"], format="%Y%m%d")
    df["season"] = df["date"].dt.year

    # create home rows
    home = df.copy()
    home["team"] = home["home_team"]
    home["opponent"] = home["visiting_team"]
    home["is_home"] = 1
    home["team_score"] = home["home_score"]
    home["opp_score"] = home["visiting_score"]

    # create away rows
    away = df.copy()
    away["team"] = away["visiting_team"]
    away["opponent"] = away["home_team"]
    away["is_home"] = 0
    away["team_score"] = away["visiting_score"]
    away["opp_score"] = away["home_score"]

    # combine
    retrosheet_base = pd.concat([home, away], ignore_index=True)

    # target
    retrosheet_base["target_win"] = (
        retrosheet_base["team_score"] > retrosheet_base["opp_score"]
    ).astype(int)

    # keep only final columns
    retrosheet_base = retrosheet_base[
        ["date", "season", "team", "opponent", "is_home", "team_score", "opp_score", "target_win"]
    ]

    return retrosheet_base
```

```python
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, confusion_matrix
)

def evaluate_model(name, y_test, y_pred, y_prob=None):
    print(f"\n===== {name} =====")

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    print("Accuracy:", acc)
    print("Precision:", prec)
    print("Recall:", rec)
    print("F1 Score:", f1)

    if y_prob is not None:
        auc = roc_auc_score(y_test, y_prob)
        print("AUC:", auc)
    else:
        auc = None

    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

    return {
        "model": name,
        "accuracy": acc,
        "precision": prec,
        "recall": rec,
        "f1": f1,
        "auc": auc
    }
```

```python
from sklearn.model_selection import cross_val_score

def run_cv(model, X, y):
    scores = cross_val_score(model, X, y, cv=5)
    print("CV scores:", scores)
    print("Mean CV:", scores.mean())
```

```python
import pandas as pd
import numpy as np

X_train_scaled = pd.read_csv("TRAIN&TEST DATASET/X_train_scaled.csv")
y_train = pd.read_csv("TRAIN&TEST DATASET/y_train.csv")
X_test_scaled = pd.read_csv("TRAIN&TEST DATASET/X_test_scaled.csv")
y_test = pd.read_csv("TRAIN&TEST DATASET/y_test.csv")

y_train = y_train.values.ravel()
y_test = y_test.values.ravel()

print(X_train_scaled.shape, y_train.shape)
print(X_test_scaled.shape, y_test.shape)
```

```python
X_train_cnn = np.array(X_train_scaled).reshape(X_train_scaled.shape[0], X_train_scaled.shape[1], 1)
X_test_cnn = np.array(X_test_scaled).reshape(X_test_scaled.shape[0], X_test_scaled.shape[1], 1)

print(X_train_cnn.shape)
print(X_test_cnn.shape)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix

def evaluate_model(name, y_test, y_pred, y_prob=None):
    print(f"\n===== {name} =====")

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    print("Accuracy:", acc)
    print("Precision:", prec)
    print("Recall:", rec)
    print("F1 Score:", f1)

    if y_prob is not None:
        auc = roc_auc_score(y_test, y_prob)
        print("AUC:", auc)
    else:
        auc = None

    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

    return {
        "model": name,
        "accuracy": acc,
        "precision": prec,
        "recall": rec,
        "f1": f1,
        "auc": auc
    }
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

model_cnn = Sequential([
    Conv1D(filters=32, kernel_size=2, activation='relu', input_shape=(X_train_cnn.shape[1], 1)),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

model_cnn.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

history = model_cnn.fit(
    X_train_cnn, y_train,
    epochs=30,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

y_prob_cnn = model_cnn.predict(X_test_cnn).ravel()
y_pred_cnn = (y_prob_cnn >= 0.5).astype(int)

result_cnn = evaluate_model("1D-CNN", y_test, y_pred_cnn, y_prob_cnn)
```
